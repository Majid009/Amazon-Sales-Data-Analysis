## Data Cleaning & Feature Engineering

To prepare the dataset for analysis, the following steps were performed:

1. **Converted the 'Date' column** to `datetime` format for easier time-based analysis.
2. **Ensured numeric columns** (`Price`, `Quantity`, `Total Sales`) were of type `int` for proper aggregation.
3. **Created new time-based features**:
   - `Month` – full month name (e.g., "January")
   - `Day` – day of the week (e.g., "Monday")
4. **Ordered months** chronologically using categorical ordering for correct visualization.

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Convert price, quantity, and total sales to integer
df['Price'] = df['Price'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['Total Sales'] = df['Total Sales'].astype(int)

# Extract month and day name
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()

# Define month order and apply
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
```

## Cohort Analysis: Customer Retention Over Time

**Objective:**
Understand customer retention behavior — how many customers made repeat purchases in the months following their first purchase.

```python
# Make sure Date is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create OrderPeriod (purchase month) and CohortMonth (first purchase month per customer)
df['OrderPeriod'] = df['Date'].dt.to_period('M')
df['CohortMonth'] = df.groupby('Customer Name')['Date'].transform('min').dt.to_period('M')

def get_cohort_index(order_period, cohort_month):
    return (order_period.dt.year - cohort_month.dt.year) * 12 + (order_period.dt.month - cohort_month.dt.month) + 1

df['CohortIndex'] = get_cohort_index(df['OrderPeriod'].astype('period[M]'), df['CohortMonth'].astype('period[M]'))


cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])['Customer Name'].nunique().reset_index()
cohort_pivot = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='Customer Name')
cohort_size = cohort_pivot.iloc[:, 0]
retention = cohort_pivot.divide(cohort_size, axis=0).round(3)

plt.figure(figsize=(10, 6))
sns.heatmap(retention, annot=True, fmt='.0%', cmap='YlGnBu')
plt.title('Customer Retention by Cohort')
plt.xlabel('Cohort Index (Months Since First Purchase)')
plt.ylabel('Cohort Month')
plt.tight_layout()
plt.show()


```

## Customer Lifetime Value (CLV)
**Objective:**
Estimate the revenue value each customer brings by combining their average order value (AOV) and purchase frequency.

```python


# Assuming df is your Amazon sales DataFrame with columns:
# 'Customer Name', 'Order ID', 'Total Sales'

# 1. Calculate total sales per customer
customer_total_sales = df.groupby('Customer Name')['Total Sales'].sum()

# 2. Calculate number of orders per customer
customer_order_counts = df.groupby('Customer Name')['Order ID'].nunique()

# 3. Calculate average order value per customer
customer_aov = customer_total_sales / customer_order_counts

# 4. Calculate purchase frequency (orders per customer)
purchase_frequency = customer_order_counts / customer_order_counts.sum()

# 5. Estimate CLV (simple model): CLV = AOV * Purchase Frequency * Total number of customers
total_customers = df['Customer Name'].nunique()
customer_clv = customer_aov * purchase_frequency * total_customers

# Put results into a DataFrame for better viewing
clv_df = pd.DataFrame({
    'Total Sales': customer_total_sales,
    'Order Count': customer_order_counts,
    'Average Order Value': customer_aov,
    'Purchase Frequency': purchase_frequency,
    'Estimated CLV': customer_clv
})

# Sort by CLV descending
clv_df = clv_df.sort_values(by='Estimated CLV', ascending=False)

# Show top 10 customers by CLV
print(clv_df.head(10))

# Plot Estimated CLV of top 10 customers
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
clv_df['Estimated CLV'].head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Customers by Estimated CLV')
plt.ylabel('Estimated Customer Lifetime Value')
plt.xlabel('Customer Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Sales Forecasting
**Objective:** Try linear regression or moving average to predict future sales.


```python
from sklearn.linear_model import LinearRegression

# Step 1: Ensure date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 2: Group by Date
daily_sales = df.groupby('Date')['Total Sales'].sum()

# Step 3: Convert index to datetime (just in case)
daily_sales.index = pd.to_datetime(daily_sales.index)

# Step 4: Prepare X and y for regression
X = daily_sales.index.map(lambda x: x.toordinal()).values.reshape(-1, 1)
y = daily_sales.values

# Step 5: Fit Linear Regression
model = LinearRegression()
model.fit(X, y)

# Step 6: Predict future sales (next 30 days)
future_dates = pd.date_range(daily_sales.index.max(), periods=30)
future_ordinals = future_dates.map(lambda x: x.toordinal()).values.reshape(-1, 1)
future_preds = model.predict(future_ordinals)

# Step 7: Plot actual and predicted
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, y, label='Actual Sales', color='blue')
plt.plot(future_dates, future_preds, label='Forecasted Sales', color='red', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Forecast (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Ensure date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 2: Group by Date
daily_sales = df.groupby('Date')['Total Sales'].sum()

# Step 3: Convert index to datetime (safety)
daily_sales.index = pd.to_datetime(daily_sales.index)

# Step 4: Prepare X and y for regression
X = daily_sales.index.map(lambda x: x.toordinal()).values.reshape(-1, 1)
y = daily_sales.values

# Step 5: Fit Linear Regression
model = LinearRegression()
model.fit(X, y)

# Step 6: Predict future sales (next 30 days)
future_dates = pd.date_range(start=daily_sales.index.max() + pd.Timedelta(days=1), periods=30)
future_ordinals = future_dates.map(lambda x: x.toordinal()).values.reshape(-1, 1)
future_preds = model.predict(future_ordinals)

# Step 7 (Extra): Calculate moving average
moving_avg = daily_sales.rolling(window=7).mean()

# Step 8 (Optional): Confidence interval (± std of residuals)
residuals = y - model.predict(X)
std_dev = np.std(residuals)
ci_upper = future_preds + std_dev
ci_lower = future_preds - std_dev

# Step 9: Plot all
plt.figure(figsize=(14, 6))
plt.plot(daily_sales.index, y, label='Actual Sales', color='blue')
plt.plot(daily_sales.index, moving_avg, label='7-Day Moving Avg', color='green', linestyle='-')
plt.plot(future_dates, future_preds, label='Forecasted Sales', color='red', linestyle='--')
plt.fill_between(future_dates, ci_lower, ci_upper, color='red', alpha=0.2, label='Confidence Interval')

# Visual markers
plt.axvline(daily_sales.index.max(), color='gray', linestyle=':', label='Forecast Start')
plt.axvline(pd.Timestamp.today(), color='black', linestyle='--', alpha=0.5, label='Today')

# Chart labels
plt.title('Sales Forecast (Linear Regression + Moving Avg + CI)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Customer Segmentation Using RFM Analysis

Objective: Segment customers based on their Recency, Frequency, and Monetary value to identify your most valuable and loyal customers.

```python
# Ensure date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 1: Compute RFM metrics
snapshot_date = df['Date'].max() + pd.Timedelta(days=1)

rfm = df.groupby('Customer Name').agg({
    'Date': lambda x: (snapshot_date - x.max()).days,   # Recency
    'Order ID': 'nunique',                              # Frequency
    'Total Sales': 'sum'                                # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

def qcut_with_dynamic_labels(series, q=5, ascending=True):
    # Get quantile-based bins, drop duplicates
    bins = pd.qcut(series, q, duplicates='drop')
    n_bins = bins.cat.categories.size  # actual number of bins
    
    # Generate labels accordingly
    if ascending:
        labels = list(range(n_bins, 0, -1))  # e.g. [5,4,3,2,1] but adjusted if fewer bins
    else:
        labels = list(range(1, n_bins + 1))
    
    # Apply qcut again with labels
    return pd.qcut(series, q, labels=labels, duplicates='drop').astype(int)

# Apply for each R, F, M with appropriate ascending order for R (recency) and descending for F and M
rfm['R_score'] = qcut_with_dynamic_labels(rfm['Recency'], q=5, ascending=True)  # Lower recency = better, so invert labels
rfm['F_score'] = qcut_with_dynamic_labels(rfm['Frequency'], q=5, ascending=False)  # Higher freq = better
rfm['M_score'] = qcut_with_dynamic_labels(rfm['Monetary'], q=5, ascending=False)  # Higher monetary = better


# Step 3: Combine RFM scores
rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Step 4: Segment customers
def segment(rfm_row):
    if rfm_row['R_score'] >= 4 and rfm_row['F_score'] >= 4:
        return 'Loyal Customer'
    elif rfm_row['R_score'] >= 4:
        return 'Recent Customer'
    elif rfm_row['F_score'] >= 4:
        return 'Frequent Buyer'
    elif rfm_row['M_score'] >= 4:
        return 'High Spender'
    else:
        return 'Other'

rfm['Segment'] = rfm.apply(segment, axis=1)

# Step 5: Plot RFM segments
plt.figure(figsize=(10, 6))
sns.countplot(data=rfm, x='Segment', order=rfm['Segment'].value_counts().index, palette='Set2', hue='Segment', legend=False)
plt.title('Customer Segments Based on RFM')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Optional: View RFM table
rfm.head()
```

### Churn Analysis: Identify and Analyze Customer Churn
What is churn?
Churn means customers who stopped buying or became inactive after some time. Detecting churn helps you understand how to retain customers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Suppose df is your sales data with at least: 'Customer ID' and 'Date' columns

# Step 1: Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 2: Define the analysis date (e.g., max date in dataset)
analysis_date = df['Date'].max()

# Step 3: Calculate recency (days since last purchase per customer)
recency_df = df.groupby('Customer Name')['Date'].max().reset_index()
recency_df['Recency'] = (analysis_date - recency_df['Date']).dt.days

# Step 4: Define churn threshold (e.g., no purchase in last 60 days = churned)
churn_threshold = 60
recency_df['Churned'] = recency_df['Recency'] > churn_threshold

# Step 5: Calculate churn rate
churn_rate = recency_df['Churned'].mean()
print(f"Churn Rate: {churn_rate:.2%}")

# Step 6: Visualize churn distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Churned', data=recency_df)
plt.title('Customer Churn Status')
plt.xlabel('Churned')
plt.ylabel('Number of Customers')
plt.xticks([0,1], ['Active', 'Churned'])
plt.show()

# Optional: Plot Recency distribution
plt.figure(figsize=(10, 6))
sns.histplot(recency_df['Recency'], bins=30, kde=True)
plt.axvline(churn_threshold, color='red', linestyle='--', label='Churn Threshold')
plt.title('Recency Distribution of Customers')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')
plt.legend()
plt.show()
```

**Insight:**
our churn rate is 27.27% for Threshold value of 60 days. It means our 27% customers did have not made a purchae in last 60 dyas.
